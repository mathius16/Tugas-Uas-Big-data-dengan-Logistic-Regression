# **logistic regression**

Disini kita akan membuat fungsi sigmoid, menghitung cost function, menghitung gradien, melakukan gradient descent, Menambahkan kolom 1 ke X untuk intercept, inisialisasi parameter theta, melakukan gradient descent, Membuat prediksi menggunakan parameter theta yang diperoleh, Menghitung akurasi model.

```python
# Fungsi sigmoid
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Fungsi untuk menghitung cost function
def cost_function(X, y, theta):
    m = len(y)
    h = sigmoid(np.dot(X, theta))
    cost = (-1 / m) * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))
    return cost

# Fungsi untuk menghitung gradien
def gradient(X, y, theta):
    m = len(y)
    h = sigmoid(np.dot(X, theta))
    grad = (1 / m) * np.dot(X.T, (h - y))
    return grad

# Fungsi untuk melakukan gradient descent
def gradient_descent(X, y, theta, alpha, num_iters):
    m = len(y)
    costs = []
    for i in range(num_iters):
        theta = theta - alpha * gradient(X, y, theta)
        cost = cost_function(X, y, theta)
        costs.append(cost)
    return theta, costs

# Menambahkan kolom 1 ke X untuk intercept
X = data[data_sub].drop('Churn', axis=1)
y = data['Churn']
X = np.c_[np.ones((len(X), 1)), X]

# Inisialisasi parameter theta
theta = np.zeros(X.shape[1])

# Melakukan gradient descent
alpha = 0.01  # Learning rate
num_iters = 1000  # Jumlah iterasi
theta, costs = gradient_descent(X, y, theta, alpha, num_iters)

# Membuat prediksi menggunakan parameter theta yang diperoleh
y_pred = sigmoid(np.dot(X, theta)) >= 0.5

# Menghitung akurasi model
accuracy = np.mean(y_pred == y) * 100
print(f"Akurasi model: {accuracy:.2f}%")
```

Disini didapatkan hasil akurasi sebesar 73.46%. Dapat kita simpulkan bahwa akurasinya sudah termasuk bagus namun belum terlalu bagus.
